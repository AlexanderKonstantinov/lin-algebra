### 1. Решить систему уравнений методом Гаусса:
$ \begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
2x_{1}+x_{2}-x_{3}+x_{4}=-2, \\
x_{1}+x_{2}-3x_{3}+x_{4}=4.
\end{cases} $  

#### Решение:

```python
# Проверка
a = np.array([[1., 1., -1.], [2., 1., -1.], [1., 1., -3.]])
b = np.array([0., -2., 4.])

c = np.linalg.solve(a, b)
print(c)
```
