## Линейные преобразования

### 1. Найти собственные векторы и собственные значения для линейного оператора, заданного матрицей
$\displaystyle A=\begin{pmatrix}
-1 & -6\\
2 & 6
\end{pmatrix}$

#### Решение:

```python
a = np.array([[-1, -6], [2, 6]])

w, v = np.linalg.eig(a)

print(f'Матрица A:\n{a}')
print(f'Собственные значения:\n{w}')
print(f'Собственные векторы:\n{v}')
```

### 2. Дан оператор поворота на 180 градусов, задаваемый матрицей
$\displaystyle A=\begin{pmatrix}
-1 & 0\\
0 & -1
\end{pmatrix}$  
Показать, что любой вектор является для него собственным.
#### Решение:

$\displaystyle \begin{vmatrix}
-1-\lambda  & 0\\
0 & -1-\lambda 
\end{vmatrix} =0$

$\displaystyle ( -1-\lambda )^{2} =0$

$\displaystyle \lambda =-1$ 

$\displaystyle \begin{cases}
-1x+0y=-1x\\
0x-1y=-1y
\end{cases}$

#### Так как x не зависит от y, и наоборот, то любой вектор для данного оператора будет являться собственным

### 3. Пусть линейный оператор задан матрицей
$\displaystyle A=\begin{pmatrix}
1 & 1\\
-1 & 3
\end{pmatrix}$  
Установить, является ли вектор  x=(1,1)  собственным вектором этого линейного оператора.
#### Решение:

$\displaystyle \begin{pmatrix}
1 & 1\\
-1 & 3
\end{pmatrix}\begin{pmatrix}
1\\
1
\end{pmatrix} =\lambda \begin{pmatrix}
1\\
1
\end{pmatrix} $  

$\displaystyle \begin{cases}
1+1=\lambda \\
-1+3=\lambda 
\end{cases} \Longrightarrow \lambda =2$  

#### Так как собственное значение определено однозначно для вектора x=(1, 1), то он будет являться собственным вектором данного линейного оператора

### 4. Пусть линейный оператор задан матрицей
$\displaystyle A=\begin{pmatrix}
0 & 3 & 0\\
3 & 0 & 0\\
0 & 0 & 3
\end{pmatrix}$  
Установить, является ли вектор  x=(3,−3,−4)  собственным вектором этого линейного оператора.
#### Решение:

$\displaystyle \begin{pmatrix}
0 & 3 & 0\\
3 & 0 & 0\\
0 & 0 & 3
\end{pmatrix}\begin{pmatrix}
3\\
-3\\
-4
\end{pmatrix} =\lambda \begin{pmatrix}
3\\
-3\\
-4
\end{pmatrix}$  

$\displaystyle \begin{cases}
0+9+0=3\lambda \\
-9+0+0=-3\lambda \\
0+0-12=-4\lambda 
\end{cases} \Longrightarrow \lambda =3$  

#### Так как собственное значение определено однозначно для вектора x=(3, -3, -4), то он будет являться собственным вектором данного линейного оператора
